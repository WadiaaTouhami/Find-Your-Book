```python
import pandas as pd
from langchain.document_loaders.csv_loader import CSVLoader
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import DataFrameLoader
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
Embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2",
                            model_kwargs={'device': 'cuda'})
```

```python
chroma_path = '/content/drive/MyDrive/chroma_books_data'
```

```python
chroma_db = Chroma(persist_directory=chroma_path, embedding_function=Embeddings)
chroma_db
```

```python
q1 = "Books exploring the impact of technology on human relationships"

q2 = "Novels that delve into the psychology of decision-making under pressure"

q3 = "Non-fiction works about the evolution of language and communication"

q4 = "Stories that examine the concept of identity in multicultural societies"

q5 = "Books discussing the ethical implications of artificial intelligence"

q6 = "Literary works that explore the theme of time and memory"

q7 = "Texts analyzing the role of mythology in modern storytelling"

q8 = "Books about the intersection of art, science, and philosophy"

q9 = "Narratives that challenge traditional notions of success and happiness"

q10 = "Works examining the psychological effects of social media on youth"
```

```python
k=5
relevant_docs_q1 = chroma_db.similarity_search(q1, k=k)
relevant_docs_q2 = chroma_db.similarity_search(q2, k=k)
relevant_docs_q3 = chroma_db.similarity_search(q3, k=k)
relevant_docs_q4 = chroma_db.similarity_search(q4, k=k)
relevant_docs_q5 = chroma_db.similarity_search(q5, k=k)
relevant_docs_q6 = chroma_db.similarity_search(q6, k=k)
relevant_docs_q7 = chroma_db.similarity_search(q7, k=k)
relevant_docs_q8 = chroma_db.similarity_search(q8, k=k)
relevant_docs_q9 = chroma_db.similarity_search(q9, k=k)
relevant_docs_q10 = chroma_db.similarity_search(q10, k=k)
```

```python
def print_results(relevant_docs, q):
  print(q)
  print()
  for i in range(k):
    title = relevant_docs[i].metadata["title"]
    print(f"Book {i + 1}:")
    print()
    print(f" - Title: {relevant_docs[i].metadata['title']}")
    print()
    print(f" - Description: {relevant_docs[i].page_content}")
    print("\n")
```

```python
print_results(relevant_docs_q1, q1)
```
